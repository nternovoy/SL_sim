```python
#Responsing to users' actions
from flask_ngrok import run_with_ngrok
from flask import Flask, request, make_response, Response
import os
import json



app = Flask(__name__)
# run_with_ngrok(app)  # Start ngrok when app is run
@app.route("/")
def hello():
    return "Hello World!"


@app.route("/hello")
def hello_there():
    return "General Kenobi"

@app.route("/slack/message_actions", methods=["POST"])
def message_actions():

    # Parse the request payload
    form_json = json.loads(request.form["payload"])
    print(form_json)

    
    
@app.route("/", methods=["POST"])
def root_method():
    # Parse the request payload
    form_json = json.loads(request.form["payload"])
    print(form_json)
    selection = form_json["actions"][0]["value"]
    print(selection)
    
    if selection == "war":
        sc.api_call(
        "chat.postMessage"
        , content_type = "application/json"
        , channel="dogs-slack"
        , text = "This game is a bit dangerous, you know"
        , attachments = '''
        [
        {
            "text": "How about a cup of coffee instead?",
            "fallback": "You are unable to choose a game",
            "callback_id": "wopr_game_coffee",
            "color": "#3AA3E3",
            "attachment_type": "default",
            "actions": 
            [
                {
                    "name": "game",
                    "text": "Okay",
                    "type": "button",
                    "value": "okay"
                },
                {
                    "name": "game",
                    "text": "NO GET OUT OF MY WAY",
                    "type": "button",
                    "value": "no"
                }
            ]
            }
            ]'''
            )
    elif selection == "chess":
        sc.api_call(
        "chat.postMessage"
        , content_type = "application/json"
        , channel="dogs-slack"
        , text = "Nope that's boring"
        )
    elif selection == "maze":
        sc.api_call(
        "chat.postMessage"
        , content_type = "application/json"
        , channel="dogs-slack"
        , text = "Nice!"
        , attachments = '''
        [
        {
            "text": "Where you wanna go first?",
            "fallback": "You are unable to choose a game",
            "callback_id": "maze_choice",
            "color": "#3AA3E3",
            "attachment_type": "default",
            "actions": 
            [
                {
                    "name": "game",
                    "text": "Right",
                    "type": "button",
                    "value": "right"
                },
                {
                    "name": "game",
                    "text": "Left",
                    "type": "button",
                    "value": "left"
                }
            ]
            }
            ]'''
            )
        

    
    return make_response("", 200)
    
app.run()
   
```
